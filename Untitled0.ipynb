```python
import pandas as pd

```

```python
data = pd.read_csv('athlete_events.csv')
data.head(50)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter for medalists with weight data from the earliest year up to 1956
early_data = data[(data['Year'] <= 1956) & (~data['Weight'].isna()) & (~data['Medal'].isna())]

# Select sports to analyze
selected_sports = ['Basketball', 'Gymnastics', 'Swimming', 'Weightlifting', 'Boxing']

# Group by sport and year to calculate average weight
grouped = early_data[early_data['Sport'].isin(selected_sports)]
avg_weight = grouped.groupby(['Sport', 'Year'])['Weight'].mean().reset_index()

# Plot the average weight over time per sport
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_weight, x='Year', y='Weight', hue='Sport', marker='o')
plt.title('Average Weight of Medal-Winning Athletes (Earliest Year–1956)')
plt.xlabel('Year')
plt.ylabel('Average Weight (kg)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Sport')
plt.tight_layout()
plt.show()
```

```python
# Filter for medalists with height data from the earliest year up to 1956
early_data = data[(data['Year'] <= 1956) & (~data['Height'].isna()) & (~data['Medal'].isna())]

# Select sports to analyze
selected_sports = ['Basketball', 'Gymnastics', 'Swimming', 'Weightlifting', 'Boxing']
grouped = early_data[early_data['Sport'].isin(selected_sports)]

# Group by sport and year to calculate average height
avg_height = grouped.groupby(['Sport', 'Year'])['Height'].mean().reset_index()

# Compute year-to-year differences
avg_height['Height_Change'] = avg_height.groupby('Sport')['Height'].diff()

# Plot the average height over time per sport
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_height, x='Year', y='Height', hue='Sport', marker='o')
plt.title('Average Height of Medal-Winning Athletes (Earliest Year–1956)')
plt.xlabel('Year')
plt.ylabel('Average Height (cm)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Sport')
plt.tight_layout()
plt.show()
```
